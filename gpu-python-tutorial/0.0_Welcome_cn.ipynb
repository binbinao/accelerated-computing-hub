# Python GPU开发101教程

_作者：[Jacob Tomlinson](https://jacobtomlinson.dev)_

**欢迎来到Python GPU开发101教程。**

我于2019年加入NVIDIA，从那时起我开始掌握在Python中编写加速代码的基础知识。我惊讶地发现，我不需要学习C++，也不需要新的开发工具。如今在Python中编写GPU代码比以往任何时候都更容易，在本教程中，我将分享我所学到的知识，以及如何开始加速你的代码。

在本教程中，我们将涵盖：
- 什么是GPU？为什么它与CPU不同？
- CUDA开发模型概述
- Numba：Python的高性能编译器
- 在Python中编写你的第一个GPU代码
- 内存管理
- 使用pyNVML了解GPU的运行状态（内存使用率、利用率等）
- RAPIDS：一套GPU加速的数据科学库
- 在GPU上使用NumPy风格的数组
- 在GPU上使用Pandas风格的数据框
- 在GPU上执行scikit-learn风格的机器学习

参与者需要具备Python和编程概念的一般知识，但不需要GPU经验。参与者的关键收获将是了解到他们不需要做太多改变就能让代码在GPU上运行。

### 大纲

通常当本教程作为会议的一部分运行时，会为参与者提供计算平台。这通常是一个定制的[Binder](https://mybinder.org/)平台，提供GPU并预装了驱动程序等依赖项。如果你现在正在会议上跟随本教程，请向你的讲师咨询平台详情。

或者，你可以使用自己的环境。所有研讨会材料都与满足[RAPIDS系统要求](https://docs.rapids.ai/install/)的系统兼容。如果你的系统满足这些要求，你可以使用conda、pip或docker按照[RAPIDS安装说明](https://docs.rapids.ai/install/)安装RAPIDS。

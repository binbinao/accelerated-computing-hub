**1. 创建一个包含从 `1` 到 `500_000_000` 的数字的输入数据数组。**

**6. 创建一个包含 `5_000_000` 个元素的新数组，表示从 `0` 到 `1000` 的线性空间。**

```python
arr = cp.linspace(0,1000, num=5_000_000)
arr
```

**7. 创建一个 `10_000` 行 `5_000` 列的随机数组。**

```python
arr = cp.random.random((10_000, 5_000))
arr
```

**8. 对该数组进行排序。**

```python
arr = cp.sort(arr)
arr
```

**附加题 9. 将数组重塑为其中一个维度长度为 `5`**

```python
arr = arr.reshape(10_000, 1_000, 5)
arr
```
